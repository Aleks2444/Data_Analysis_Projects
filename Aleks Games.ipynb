# Base Code

```python
import random

def attack(health, opp_health):
    player_damage = min(random.randint(0, 20), opp_health)  # Cap damage to player's remaining health
    opp_damage = min(random.randint(0, 20), health)  # Cap damage to opponent's remaining health
    
    opp_health -= player_damage
    health -= opp_damage
    
    health = max(0, health)
    opp_health = max(0, opp_health)
    
    return health, opp_health, player_damage, opp_damage

def user_only_attack(opp_health):
    player_damage = min(random.randint(0, 20), opp_health)  # Cap damage to opponent's remaining health
    opp_health -= player_damage
    opp_health = max(0, opp_health)  # Ensure opponent's health doesn't go negative
    
    return opp_health, player_damage
    
def cpu_only_attack(health):
    opp_damage = min(random.randint(0, 20), health)  # Cap damage to users remaining health
    health -= opp_damage
    health = max(0, health)
    
    return health, opp_damage
    
    
def block(health, opp_health):
    return health, opp_health, 0, 0

def heal(health):
    player_heal = random.randint(0, 15)
    
    health += player_heal
    health = min(100, health)  # Cap health to 100
    
    return health, player_heal

def cpu_heal(opp_health):
    opp_heal = random.randint(0, 15)
    
    opp_health += opp_heal
    opp_health = min(100, opp_health)  # Cap health to 100
    
    return opp_health, opp_heal
    
def main():
    health = 100
    opp_health = 100

    while health > 0 and opp_health > 0:
        print(f"Your health: {health}, Opponent's health: {opp_health}")

        decision = input("Would you like to (A): Attack, (B): Block, (H): Heal? ").lower()

        # Computer's decision: 50% chance to attack, 30% chance to block and 20% chance to heal
        comp_decision = random.choices(["a", "b", "h"], weights=[0.5, 0.3, 0.2])[0]

        print(f"Computer chose: {comp_decision.upper()}")

        if decision == "a" and comp_decision == "a":
            print("You have decided to attack")
            print("Your opponent attacked you.")
            health, opp_health, player_damage, opp_damage = attack(health, opp_health)
            print(f"You dealt {player_damage} damage to your opponent.")
            print(f"You took {opp_damage} damage.")

        elif decision == "b" and comp_decision == "a":
            print("You have decided to block")
            print("Your opponent attacked you.")
            if random.random() < 0.2:
                damage_taken = random.randint(1, 15)
                opp_health -= damage_taken
                print(f"Your opponent took {damage_taken} damage as their sword ricocheted off your shield and onto their shin.")
            else:
                print("You successfully blocked the attack.")
                
        elif decision == "h" and comp_decision == "a":
            health, player_heal = heal(health)
            print(f"You heal for: {player_heal} damage")
            health, opp_damage = cpu_only_attack(opp_health)
            print(f"Your opponent attacks you while you are healing and does {opp_damage}")    
            
        elif decision == "a" and comp_decision == "b":
            print("You have decided to attack")
            print("Your opponent blocked.")
            if random.random() < 0.2:
                damage_taken = random.randint(1, 15)
                health -= damage_taken
                print(f"Your sword bounced off the opponent's shield, and you hit yourself, taking {damage_taken} damage.")
            else:
                print("Your opponent successfully blocked your attack.")

        elif decision == "b" and comp_decision == "b":
            print("You have decided to block")
            print("Your opponent blocked.")
            # Nothing happens when both block
            
        elif decision == "h" and comp_decision == "b":
            health, player_heal = heal(health)
            print(f"Your opponent anticipates a block but you chose to heal and restored {player_heal} damage")
            
        elif decision == "a" and comp_decision == "h":
            opp_health, opp_heal = cpu_heal(opp_health)
            print(f"Your opponent chose to heal and restored {opp_heal} damage")
            opp_health, player_damage = user_only_attack(opp_health)
            print(f"Meanwhile you chose to attack and dealt {player_damage} to your opponent")
            
        elif decision == "b" and comp_decision == "h":
            opp_health, opp_heal = cpu_heal(opp_health)
            print(f"You chose to block but your opponent went to heal themselves and restored {opp_heal} damage")
            
        elif decision == "h" and comp_decision == "h":
            opp_health, opp_heal = cpu_heal(opp_health)
            health, user_heal = heal(health)
            print(f"You healed for {user_heal} damage while your opponent healed themselves for {opp_heal} damage")
            
        print("""
        """)
            
            
            
    if health <= 0:
        print("Game over. You lost.")
    else:
        print("Congratulations! You defeated your opponent!")

if __name__ == "__main__":
    main()
```
